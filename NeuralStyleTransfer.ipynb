# Neural Style Transfer

## 0. Import dependencies and a pre-trained neural style transfer model from Tensorflow Hub

```python
import tensorflow_hub as hub
import tensorflow as tf
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
# load pretrained model
model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

## 1. Tensorflow preprocessing: loading style and images

```python
def load_image(img_path):
    """
    This is used to load up and preprocess images using 
    tensorflow.  Final returns the processed image.
    """
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img, channels=3) # decoding image
    img = tf.image.convert_image_dtype(img, tf.float32) # reformat image data type
    img = img[tf.newaxis, :] # Add image within a new array, ":" will grab the entire image
    return img
```

```python
# load images for content and style
content_image = load_image('eisaku.png')
style_image = load_image('monet.jpeg')
```

## 3. Stylize image using neural sytle transfer

```python
# use the loaded model in the beginning
stylized_image = model(tf.constant(content_image), tf.constant(style_image))[0]
```

```python
plt.imshow(np.squeeze(stylized_image))
plt.show()
```

```python
cv2.imwrite('generated_img.jpg', cv2.cvtColor(np.squeeze(stylized_image)*255, cv2.COLOR_BGR2RGB))
```
